This GUI is very basic and not efficient in the slightest, needs lots of work and was just a basic way of analysing the model's performance.

```python
import cv2
import mediapipe as mp
import numpy as np
from joblib import load
import sklearn
import tkinter as tk
from PIL import Image, ImageTk

# Load the trained SVM model
model = load('C:/Users/ashnr/Downloads/Updated_svm_model.joblib')

# Initialize MediaPipe Pose
mp_pose = mp.solutions.pose
pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# Define the indexes of landmarks to track
landmark_indexes = [0, 11, 12, 23, 24, 25, 26, 27, 28, 15, 16, 13, 14]

# Define connections between landmarks using the indices specified in 'landmark_indexes'
connections = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 4), (3, 5), (4, 6), (5, 7), (6, 8), (1, 12), (12, 10), (2, 11), (11, 9)]

# Define URL of video if testing model on videos or IP address URL if testing model on camera using DroidCam
#stream_url = 'http://192.168.1.128:4747/video'
stream_url = "C:/Users/ashnr/Desktop/GTA Clips/test.mp4"

# Initialize the video capture with the stream URL
cap = cv2.VideoCapture(stream_url)

# Create Tkinter GUI called Fall Detection Application
fall_app = tk.Tk()
fall_app.title("Fall Detection Application")

# Create main video stream label
video = tk.Label(fall_app, borderwidth=5, relief="solid")
video.grid(row = 0, column = 0, rowspan = 10, padx=20, pady=20, sticky="NSEW")

# Create title label
title = tk.Label(fall_app, text="Fall Detection", font = ("Arial", 40), height = "1", borderwidth=1, relief="solid")
title.grid(row = 0, column = 1, sticky="NSEW", ipadx=100, padx = 20, pady = 20, columnspan=4)

# Create prediction label
prediction_label = tk.Label(fall_app, text = "No Prediction", font = ("Arial", 25), borderwidth=1, relief="solid", bg="white")
prediction_label.grid(row = 1, column=1, sticky="NSEW", padx = 20, pady = 20, columnspan=4)

# Create pose landmarks label
landmarks_label = tk.Label(fall_app, text = "Coordinate Landmarks", font = ("Arial", 15), borderwidth=1, relief="solid")
landmarks_label.grid(row = 2, column=1, sticky="NSEW", padx = 20, pady = 20, columnspan=4)

# Create label for each landmark's x and y
x_nose = tk.Label(fall_app, text = "X Nose:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_nose.grid(row = 3, column=1, sticky="NSEW", padx = 20, pady = 20)

y_nose = tk.Label(fall_app, text = "Y Nose:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_nose.grid(row = 3, column=2, sticky="NSEW", padx = 20, pady = 20)

x_left_shoulder = tk.Label(fall_app, text = "X Left Shoulder:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_left_shoulder.grid(row = 3, column=3, sticky="NSEW", padx = 20, pady = 20)

y_left_shoulder = tk.Label(fall_app, text = "Y Left Shoulder:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_left_shoulder.grid(row = 3, column=4, sticky="NSEW", padx = 20, pady = 20)

x_right_shoulder = tk.Label(fall_app, text = "X Right Shoulder:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_right_shoulder.grid(row = 4, column=1, sticky="NSEW", padx = 20, pady = 20)

y_right_shoulder = tk.Label(fall_app, text = "Y Right Shoulder:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_right_shoulder.grid(row = 4, column=2, sticky="NSEW", padx = 20, pady = 20)

x_left_hip = tk.Label(fall_app, text = "X Left Hip:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_left_hip.grid(row = 4, column=3, sticky="NSEW", padx = 20, pady = 20)

y_left_hip = tk.Label(fall_app, text = "Y Left Hip:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_left_hip.grid(row = 4, column=4, sticky="NSEW", padx = 20, pady = 20)

x_right_hip = tk.Label(fall_app, text = "X Right Hip:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_right_hip.grid(row = 5, column=1, sticky="NSEW", padx = 20, pady = 20)

y_right_hip = tk.Label(fall_app, text = "Y Right Hip:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_right_hip.grid(row = 5, column=2, sticky="NSEW", padx = 20, pady = 20)

x_left_knee = tk.Label(fall_app, text = "X Left Knee:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_left_knee.grid(row = 5, column=3, sticky="NSEW", padx = 20, pady = 20)

y_left_knee = tk.Label(fall_app, text = "Y Left Knee:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_left_knee.grid(row = 5, column=4, sticky="NSEW", padx = 20, pady = 20)

x_right_knee = tk.Label(fall_app, text = "X Right Knee:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_right_knee.grid(row = 6, column=1, sticky="NSEW", padx = 20, pady = 20)

y_right_knee = tk.Label(fall_app, text = "Y Right Knee:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_right_knee.grid(row = 6, column=2, sticky="NSEW", padx = 20, pady = 20)

x_left_foot = tk.Label(fall_app, text = "X Left Foot:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_left_foot.grid(row = 6, column=3, sticky="NSEW", padx = 20, pady = 20)

y_left_foot = tk.Label(fall_app, text = "Y Left Foot:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_left_foot.grid(row = 6, column=4, sticky="NSEW", padx = 20, pady = 20)

x_right_foot = tk.Label(fall_app, text = "X Right Foot:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_right_foot.grid(row = 7, column=1, sticky="NSEW", padx = 20, pady = 20)

y_right_foot = tk.Label(fall_app, text = "Y Right Foot:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_right_foot.grid(row = 7, column=2, sticky="NSEW", padx = 20, pady = 20)

x_left_wrist = tk.Label(fall_app, text = "X Left Wrist:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_left_wrist.grid(row = 7, column=3, sticky="NSEW", padx = 20, pady = 20)

y_left_wrist = tk.Label(fall_app, text = "Y Left Wrist:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_left_wrist.grid(row = 7, column=4, sticky="NSEW", padx = 20, pady = 20)

x_right_wrist = tk.Label(fall_app, text = "X Right Wrist:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_right_wrist.grid(row = 8, column=1, sticky="NSEW", padx = 20, pady = 20)

y_right_wrist = tk.Label(fall_app, text = "Y Right Wrist:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_right_wrist.grid(row = 8, column=2, sticky="NSEW", padx = 20, pady = 20)

x_left_elbow = tk.Label(fall_app, text = "X Left Elbow:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_left_elbow.grid(row = 8, column=3, sticky="NSEW", padx = 20, pady = 20)

y_left_elbow = tk.Label(fall_app, text = "Y Left Elbow:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_left_elbow.grid(row = 8, column=4, sticky="NSEW", padx = 20, pady = 20)

x_right_elbow = tk.Label(fall_app, text = "X Right Elbow:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
x_right_elbow.grid(row = 9, column=2, sticky="NSEW", padx = 20, pady = 20)

y_right_elbow = tk.Label(fall_app, text = "Y Right Elbow:\n", font = ("Arial", 10), borderwidth=1, relief="solid")
y_right_elbow.grid(row = 9, column=3, sticky="NSEW", padx = 20, pady = 20)


# Define start stream function. Creates video stream, gets pose detection landmark on video, makes prediction using model and sets landmark values labels, sets prediction label and adds skeleton to video.
def start_stream():

    prediction_text = ""
    prediction = []
    bg_color = "white"
    ret, frame = cap.read()

    # Convert the BGR image to RGB, flip the image for later annotation
    frame = cv2.cvtColor(cv2.flip(frame, 1), cv2.COLOR_BGR2RGB)
    frame.flags.writeable = False

    #frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
    frame = cv2.flip(frame,1)

    # Detect pose landmarks
    results = pose.process(frame)

    # Draw pose landmarks on the frame for visualization
    frame.flags.writeable = True
    frame = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

    if results.pose_landmarks:
        # Extract selected landmarks
        selected_pose_landmarks = [results.pose_landmarks.landmark[i] for i in landmark_indexes]
        pose_landmarks = [[lmk.x, lmk.y] for lmk in selected_pose_landmarks]
        pose_landmarks_flat = np.array(pose_landmarks).flatten().reshape(1, -1)

        # Predict fall/no fall
        prediction = model.predict(pose_landmarks_flat)

        if prediction[0] == 1:
            prediction_text = "No Fall Detected!"
            skeleton_color = (0, 255, 0)
            bg_color = "green"
        else:
            prediction_text = "Fall Detected!"
            skeleton_color = (0, 0, 255)
            bg_color = "red"

        # Manually draw specified landmarks and connections
        for landmark in selected_pose_landmarks:
            x, y = int(landmark.x * frame.shape[1]), int(landmark.y * frame.shape[0])
            cv2.circle(frame, (x, y), 5, skeleton_color, -1)

        # Draw the specified connections
        for connection in connections:
            start_idx, end_idx = connection
            start_landmark = selected_pose_landmarks[start_idx]
            end_landmark = selected_pose_landmarks[end_idx]
            start_x, start_y = int(start_landmark.x * frame.shape[1]), int(start_landmark.y * frame.shape[0])
            end_x, end_y = int(end_landmark.x * frame.shape[1]), int(end_landmark.y * frame.shape[0])
            cv2.line(frame, (start_x, start_y), (end_x, end_y), skeleton_color, 2)

    # Change colours of video frame to RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Convert frame to array
    conv_frame = Image.fromarray(frame)

    # Convery array to image
    photo_image = ImageTk.PhotoImage(image = conv_frame)

    # Set video label to image frame
    video.photo_image = photo_image

    # Update video label
    video.configure(image = photo_image)

    # Do previous function every 5ms
    video.after(5, start_stream)

    # Update prediction label
    prediction_label.configure(text = prediction_text, bg = bg_color)

    # Function that updates all the landmark labels
    def landmarks_configure():
        x_nose.configure(text = ("X Nose:\n" + f"{selected_pose_landmarks[0].x:.4f}"))
        x_left_shoulder.configure(text = ("X Left Shoulder:\n" + f"{selected_pose_landmarks[2].x:.4f}"))
        x_right_shoulder.configure(text = ("X Right Shoulder:\n" + f"{selected_pose_landmarks[1].x:.4f}"))
        x_left_hip.configure(text = ("X Left Hip:\n" + f"{selected_pose_landmarks[3].x:.4f}"))
        x_right_hip.configure(text = ("X Right Hip:\n" + f"{selected_pose_landmarks[4].x:.4f}"))
        x_left_knee.configure(text = ("X Left Knee:\n" + f"{selected_pose_landmarks[5].x:.4f}"))
        x_right_knee.configure(text = ("X Right Knee:\n" + f"{selected_pose_landmarks[6].x:.4f}"))
        x_left_foot.configure(text = ("X Left Foot:\n" + f"{selected_pose_landmarks[7].x:.4f}"))
        x_right_foot.configure(text = ("X Right Foot:\n" + f"{selected_pose_landmarks[8].x:.4f}"))
        x_left_wrist.configure(text = ("X Left Wrist:\n" + f"{selected_pose_landmarks[9].x:.4f}"))
        x_right_wrist.configure(text = ("X Right Wrist:\n" + f"{selected_pose_landmarks[10].x:.4f}"))
        x_left_elbow.configure(text = ("X Left Elbow:\n" + f"{selected_pose_landmarks[11].x:.4f}"))
        x_right_elbow.configure(text = ("X Right Elbow:\n" + f"{selected_pose_landmarks[12].x:.4f}"))

        y_nose.configure(text = ("Y Nose:\n" + f"{selected_pose_landmarks[0].y:.4f}"))
        y_left_shoulder.configure(text = ("Y Left Shoulder:\n" + f"{selected_pose_landmarks[2].y:.4f}"))
        y_right_shoulder.configure(text = ("Y Right Shoulder:\n" + f"{selected_pose_landmarks[1].y:.4f}"))
        y_left_hip.configure(text = ("Y Left Hip:\n" + f"{selected_pose_landmarks[3].y:.4f}"))
        y_right_hip.configure(text = ("Y Right Hip:\n" + f"{selected_pose_landmarks[4].y:.4f}"))
        y_left_knee.configure(text = ("Y Left Knee:\n" + f"{selected_pose_landmarks[5].y:.4f}"))
        y_right_knee.configure(text = ("Y Right Knee:\n" + f"{selected_pose_landmarks[6].y:.4f}"))
        y_left_foot.configure(text = ("Y Left Foot:\n" + f"{selected_pose_landmarks[7].y:.4f}"))
        y_right_foot.configure(text = ("Y Right Foot:\n" + f"{selected_pose_landmarks[8].y:.4f}"))
        y_left_wrist.configure(text = ("Y Left Wrist:\n" + f"{selected_pose_landmarks[9].y:.4f}"))
        y_right_wrist.configure(text = ("Y Right Wrist:\n" + f"{selected_pose_landmarks[10].y:.4f}"))
        y_left_elbow.configure(text = ("Y Left Elbow:\n" + f"{selected_pose_landmarks[11].y:.4f}"))
        y_right_elbow.configure(text = ("Y Right Elbow:\n" + f"{selected_pose_landmarks[12].y:.4f}"))


    # Runs landmark configure function
    landmarks_configure()

# Runs start stream function
start_stream()

# Create function that closes GUI
def close_win(e):
   fall_app.destroy()

# Bind close GUI function to escape key. If pressed, closes window
fall_app.bind('<Escape>', lambda e: close_win(e))

fall_app.mainloop()

cap.release()
```
